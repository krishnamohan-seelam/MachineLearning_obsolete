```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import matplotlib.cm as cm
```

```python
suicides_dataset_2013 = pd.read_csv('Suicide/Suicides_by_causes_state_2013.csv',encoding = "ISO-8859-1")
suicides_dataset_2001_2012 = pd.read_csv('Suicide/Suicides_by_causes_state.csv',encoding = "ISO-8859-1")
```

```python
suicides_dataset_2013.head(5)
```

```python
suicides_dataset_2001_2012.head(5)
```

```python
#Since there are 29 states and 10 UT ,I will at analyze national level. This would gives an insight on the overall data
suicides_dataset =suicides_dataset_2001_2012.append(suicides_dataset_2013)
```

```python
plt.style.use('seaborn-colorblind')
overall = suicides_dataset[(suicides_dataset['STATE/UT'] =='TOTAL (ALL INDIA)' )
                                   &(suicides_dataset['CAUSE']=='Total')][['Year','Grand Total']] 

fig, ax = plt.subplots(figsize=(6, 4))
plt.plot(overall['Year'], overall['Grand Total'])
ax.set_title('Overall suicide trend in years (2001 -2013)')
ax.set_ylabel('Number of suicides')
ax.set_xlabel('Year')
plt.show()
```

```python
# On data observation its found that it contains 
suicides_causes = suicides_dataset[(suicides_dataset['STATE/UT'] =='TOTAL (ALL INDIA)' )
                                   &(suicides_dataset['CAUSE']!='Total')
                                   &(suicides_dataset['CAUSE']!='Total Illness')
                                 ]
```

```python
#fig, ax = plt.subplots(figsize=(8, 5))
plt.style.use('seaborn-colorblind')
year_wise =suicides_causes.groupby('Year').agg({ 'Total Female':np.sum,'Total Male':np.sum,})
 
ax=year_wise.plot.barh(width=0.75)
ax.grid(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_title('Male vs Female Suicide Comparison')
fig = ax.figure
fig.set_size_inches(8, 6)
```

```python
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10,8),nrows=1, ncols=2 ,)
 
age_wise_m =suicides_causes.groupby('Year').agg({'Male upto 14 years':np.sum, 
                                                    
                                                    'Male 15-29 years':np.sum,
                                                    
                                                    'Male 30-44 years': np.sum,
                                                    
                                                    'Male 45-59 years': np.sum,
                                                    
                                                    'Male 60 years and above': np.sum,
                                                     
                                                   })
ax1=age_wise_m.plot(ax=axes[0])
ax1.grid(False)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.legend(loc='upper left', frameon = False,fontsize ='x-small') 
#ax1.set_fontsize('6')
ax1.set_ylabel('Number of suicides')
ax1.patch.set_facecolor('white')
ax1.set_title('Males- Who is at risk')


age_wise_f =suicides_causes.groupby('Year').agg({'Female upto 14 years':np.sum, 
                                                    
                                                    'Female 15-29 years':np.sum,
                                                    
                                                    'Female 30-44 years': np.sum,
                                                    
                                                    'Female 45-59 years': np.sum,
                                                    
                                                    'Female 60 years and above': np.sum,
                                                     
                                                   })
ax2=age_wise_f.plot(ax=axes[1])
ax2.grid(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.legend(loc='upper left', frameon = False,fontsize ='x-small') 
ax2.set_title('Females- Who is at risk')
#ax2.set_ylabel('Number of suicides')
ax2.patch.set_facecolor('white')
plt.tight_layout()
plt.show()
fig.savefig('Assignment04.png', dpi=fig.dpi)
```

```python
suicides_causes_2001 = suicides_causes[(suicides_causes['STATE/UT'] =='TOTAL (ALL INDIA)' )
                                       &(suicides_causes['CAUSE']!='Total')
                                       &(suicides_causes['CAUSE']!='Total Illness')
                                       &(suicides_causes['Year'] == 2001 )
                                    ]
cause_male = suicides_causes_2001.groupby('CAUSE').agg({'Total Male' :np.sum})['Total Male'].nlargest(10)
 
cause_female = suicides_causes_2001.groupby('CAUSE').agg({'Total Female' :np.sum})['Total Female'].nlargest(10)
 
```

```python
fig, axes = plt.subplots(figsize=(12,8),nrows=1, ncols=2,sharex=True)
ax1=cause_male.plot.bar(ax=axes[0],alpha=0.9,color='#2196F3',rot = 45)
ax1.grid(False)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.legend(loc='best', frameon = False,fontsize ='small') 
#ax1.set_fontsize('6')
ax1.set_title(" Top 10 causes of male suicides-2001")
ax1.set_xlabel('')
ax1.set_ylabel('Number of suicides')
ax1.set_xticklabels(ax1.get_xticklabels(),ha='right')

ax2=cause_female.plot.bar(ax=axes[1],alpha=0.9, color='#E91E63',rot = 45)
ax2.grid(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.legend(loc='best', frameon = False,fontsize ='small') 
ax2.set_xlabel('')
ax2.set_title(" Top 10 causes of female suicides-2001")
ax2.set_xticklabels(ax1.get_xticklabels(),ha='right')
plt.tight_layout()
plt.show()
```

```python
suicides_causes_2013 = suicides_causes[(suicides_causes['STATE/UT'] =='TOTAL (ALL INDIA)' )
                                       &(suicides_causes['CAUSE']!='Total')
                                       &(suicides_causes['CAUSE']!='Total Illness')
                                       &(suicides_causes['Year'] == 2013 )]
cause_male = suicides_causes_2013.groupby('CAUSE').agg({'Total Male' :np.sum})['Total Male'].nlargest(10)
 
cause_female = suicides_causes_2013.groupby('CAUSE').agg({'Total Female' :np.sum})['Total Female'].nlargest(10)                                       
```

```python
fig, axes = plt.subplots(figsize=(12,8),nrows=1, ncols=2,sharex=True)
 
ax1=cause_male.plot.bar(ax=axes[0],alpha=0.9,color='#2196F3',rot = 45)
ax1.grid(False)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.legend(loc='best', frameon = False) 
#ax1.set_fontsize('6')
ax1.set_title(" Top 10 causes of male suicides-2013")
ax1.set_ylabel('')
ax1.set_xticklabels(ax1.get_xticklabels(),ha='right')

ax2=cause_female.plot.bar(ax=axes[1],alpha=0.9, color='#E91E63',rot =45)
ax2.grid(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.legend(loc='best', frameon = False) 
ax2.set_ylabel('')
ax2.set_title(" Top 10 causes of female suicides-2013")
ax2.set_xticklabels(ax1.get_xticklabels(),ha='right')
plt.tight_layout()
plt.show()
```

```python
cause_total = suicides_causes.groupby('CAUSE').agg({'Grand Total' :np.sum})['Grand Total'].nlargest(10)

fig, axes = plt.subplots(figsize=(8,6),nrows=1, ncols=1,sharex=True)
#cmap = cm.tab10
#colors = cmap(np.linspace(0., 1.,10))
colors_10 = ['#f44336', '#E91E63', '#9C27B0', '#673AB7' ,'#03A9F4','#4CAF50','#FFEB3B','#FF9800','#795548','#84FFFF']
ax1=cause_total.plot.pie(ax=axes,autopct='%.2f',colors=colors_10 )
ax1.grid(False)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
 
ax1.set_title("Top 10 causes of suicides:2001-2013")
ax1.set_ylabel('')


plt.tight_layout()
plt.show()
```
